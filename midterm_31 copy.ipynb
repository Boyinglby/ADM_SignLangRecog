```python
import pandas as pd
import glob
import os

# Specify the path 
folder_path = 'data/'  
num_points = 20
coordinates = ['x', 'y', 'z']

# Generate column names: p1_x, p1_y, p1_z, p2_x, ..., p20_z
column_names = [f'p{point}_{coord}' for point in range(0, num_points + 0) for coord in coordinates]

# Verify the number of columns
assert len(column_names) == 60
# Define a function to process files


file_pattern = folder_path + 'bye_*_[9].txt'
file_list = glob.glob(file_pattern)

# Group files by their prefix
file_groups = {}
for file in file_list:
    # Extract the prefix (e.g., 'bye_x', 'bye_y', ...)
    prefix = '_'.join(os.path.basename(file).split('_')[:-1])
    if prefix not in file_groups:
        file_groups[prefix] = []
    file_groups[prefix].append(file)
    print(prefix)

# Initialize dictionaries to store means and stds
means = {}
stds = {}
combined_df = []
# Process each group of files
for prefix, files in file_groups.items():
    dataframes = []
    for file in files:
        df = pd.read_csv(file, 
                 delimiter=r'\s+',  # Use regex to handle any whitespace
                 header=None,       # No header in the file
                 names=column_names)  # Assign the generated column names
    dataframes.append(df)

    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Calculate mean and std for each column
    means[prefix] = combined_df.mean()
    stds[prefix] = combined_df.std()

# Convert means and stds dictionaries to DataFrames
means_df = pd.DataFrame(means)
stds_df = pd.DataFrame(stds)

# Calculate mean of means and mean of stds for each column
mean_of_means = means_df.mean(axis=1)
mean_of_stds = stds_df.mean(axis=1)
std_of_stds = stds_df.std(axis=1)
file_groups
```
